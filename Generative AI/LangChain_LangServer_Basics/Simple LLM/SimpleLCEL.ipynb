```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.1-8b-instant",groq_api_key=groq_api_key)
```

```python
from langchain_core.messages import HumanMessage,SystemMessage
```

```python
## System message is telling the LLM model on how to behave 
messages = [
    SystemMessage(content="Translate the following text from english to Spanish"),
    HumanMessage(content="You are late for class")
]
```

```python
aiResponse = model.invoke(messages)
```

```python
## To parse the messages coming from the LLM model we a output parser
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
```

```python
parser.invoke(aiResponse)
```

```python
### Using LCEL we can change the components
chain = model|parser
chain.invoke(messages)
```

```python
##Prompt templates
from langchain_core.prompts import ChatPromptTemplate
## Using generic template via placeholders to make it reusable for any language
generic_template = "Translate the following text from english to:  {language}"
prompt=ChatPromptTemplate.from_messages([("system",generic_template),("human","{text}")])
```

```python
result=prompt.invoke({"language":"Spanish","text":"You are late for class"})
```

```python
result.to_messages()
```

```python
##LCEL allows us to change the components in the chain without changing the code
chain=prompt|model|parser
chain.invoke({"language":"German","text":"You are late for class"})
```
